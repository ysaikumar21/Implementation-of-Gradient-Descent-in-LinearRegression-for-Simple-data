```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,3,4,5])
y_train = np.array([5,7,9,11,13])

num_of_rows = len(x_train)

m_value = 0
c_value = 0

iterations = 1000
learning_rate = 0.05
plt.scatter(x_train,y_train,color='r')
for i in range(iterations):
  y_train_pred = m_value * x_train + c_value

  plt.plot(x_train,y_train_pred,color='g')
  cost_fun = (1/num_of_rows) * sum([i**2 for i in (y_train-y_train_pred)])
  m_der = -(2/num_of_rows)*sum(x_train * (y_train-y_train_pred))
  c_der = -(2/num_of_rows)*sum(y_train,y_train_pred)

  m_value = m_value - learning_rate * m_der
  c_value = c_value -learning_rate * c_der

  print(f'm : {m_value} and c : {c_value} iteration : {i}')

```
